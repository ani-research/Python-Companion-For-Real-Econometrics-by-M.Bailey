```python
DonutDF= rdf["dta"]
display(DonutDF)
```

* To make frequency table:

```python
freq = (
    DonutDF["malex"]                       # From the column malex (this is now a series(single column))
    .value_counts()                        # Give me the frequency table, this counts the number of numbers in the column and makes a table
    .sort_index()                          # Now if number of 1s in more than 0 then in the table we will see number of 1s above 0s so this command makes sure that 0 is always above 1
# This is now a series which we will convert to dataframe, a series has an index and one column
# Right now we have a series with index = unique vales of malex and column with our count values
    .rename_axis("Value")                  # Change the index name to "value"
    .reset_index(name="Observations")      # This shifts index into becoming a column, and inserts a new index which is just 0 1 2 and so on. The name parameter names the counts column as "observations" (default is the name "count")
)

print(freq)
```

```python
fig, ax = plt.subplots()

males= DonutDF[DonutDF["malex"]==1]
# You can pass bools into df to filter them and save them in new objects 
# Here DonutDF["malex"==1] gives a boolean output, there it says True for those rows that have male = 1
# Then this is passed through DonutDF again to get the rows for which we have TRUE and then we store those rows in males

ax.scatter(males["donutsx"], males["weightx"])
plt.show()
```
